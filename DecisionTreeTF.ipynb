##Dataset

```python
train_df = pandas.read_csv("diabetes.csv")
test_df = pandas.read_csv("diabetes.csv")
print(len(test_df))
#80% for train and 20% for test
for i in range (int(len(test_df) * 0.8)):
  test_df =  test_df.drop(test_df.index[0])

#test_df = test_df.drop('Outcome', axis='columns' )

for i in range (int(len(train_df) * 0.2)):
  train_df =  train_df.drop(train_df.index[0])

print(len(test_df))
test_df.head()
#train_df.head()
```

##Feature

###Train

```python
# Convert the pandas dataframe into a TensorFlow dataset
train_ds = tfdf.keras.pd_dataframe_to_tf_dataset(train_df, label="Outcome")
# Train the model
model = tfdf.keras.RandomForestModel()
model.fit(train_ds)
```

###Teste


```python
test_df.head()
print(test_df.iloc[[21]])
```

```python
# Convert it to a TensorFlow dataset
test_ds = tfdf.keras.pd_dataframe_to_tf_dataset(test_df, label="Outcome", )

# Evaluate the model
model.compile(metrics=["accuracy"])
print(model.evaluate(test_ds))
print(model.predict(test_ds)[21])
np = train_df.to_numpy()

```
